## Stack

* Insere e Remove na Cabeça

image.png

```python
import random

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Função para inserir (push) no topo da pilha
def push(top, value):
    new_node = Node(value)
    new_node.next = top
    return new_node  # novo topo

# Função para remover (pop) do topo da pilha
def pop(top):
    if top is None:
        return None, None  # pilha vazia
    value = top.data
    temp = top
    top = top.next
    del temp  # opcional, libera memória
    return top, value

# Função para imprimir a pilha
def print_stack(top):
    current = top
    while current:
        print(current.data)
        current = current.next

# Função para verificar se a pilha está vazia
def is_empty(top):
    return top is None

# Exemplo de uso
top = None
top = push(top, 5)
top = push(top, 7)
top = push(top, 2)

print("Pilha atual:")
print_stack(top)

top, value = pop(top)
print(f"\nPop: {value}")
print("Pilha após pop:")
print_stack(top)
```

## Queue

* Insere na Cabeça
* Remove da Cauda
* Implementação Eficiente utiliza dois ponteiros


image.png

```python
# Inicializa fila
def init_queue():
    return None, None  # head, tail

# Inserir no final (enqueue)
def enqueue(head, tail, value):
    new_node = Node(value)
    if head is None:
        head = tail = new_node
    else:
        tail.next = new_node
        tail = new_node
    return head, tail

# Remover do início (dequeue)
def dequeue(head, tail):
    if head is None:
        return None, None, None  # fila vazia
    value = head.data
    temp = head
    head = head.next
    if head is None:
        tail = None  # fila ficou vazia
    del temp
    return head, tail, value

# Imprimir fila
def print_queue(head):
    current = head
    while current:
        print(current.data)
        current = current.next

# Exemplo de uso
head, tail = init_queue()
head, tail = enqueue(head, tail, 10)
head, tail = enqueue(head, tail, 20)
head, tail = enqueue(head, tail, 30)

print("Fila FIFO atual:")
print_queue(head)

head, tail, val = dequeue(head, tail)
print(f"\nDequeue: {val}")
print("Fila FIFO após dequeue:")
print_queue(head)
```
